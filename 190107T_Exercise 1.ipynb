Question 8

```python
import matplotlib.pyplot as plt
from numpy import linalg
m ,c = 2,-4
N = 10
x = np.linspace(0, N-1, N).reshape(N,1)
sigma = 10
y = m*x + c + np.random.normal(0, sigma, (N,1))
plt.scatter(x,y)
X = np.append(np.ones((N,1)),x, axis = 1) # pre append
new_array = np.linalg.inv(X.T @ X) @ X.T @y
print(new_array)
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
im = cv.imread("gal_gaussian.png")
assert im is not None
blur_im = cv.GaussianBlur(im,(5,5),0)
blur_im2 = cv.GaussianBlur(im,(11,11),0)

#plotting the 3 images

fig, ax = plt.subplots(1,3)
fig.set_figwidth(20)
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0].set_title("gal_gadot")
ax[1].imshow(cv.cvtColor(blur_im, cv.COLOR_BGR2RGB))
ax[1].set_title("Blur_Image 1")
ax[2].imshow(cv.cvtColor(blur_im2, cv.COLOR_BGR2RGB))
ax[2].set_title("Blur_Image 2")
plt.show()
```

```python
#median blur is highly efective for salt-and-pepper noise
# cv2.medianBlur(image name,kernel_size)
#kernel size must be a positive odd integer

import cv2 as cv
import matplotlib.pyplot as plt
im = cv.imread("gal_sandp.png")
median_blur_1= cv.medianBlur(im , 3)
median_blur_2= cv.medianBlur(im , 5)

#plotting the 3 images

fig, ax = plt.subplots(1,3)
fig.set_figwidth(20)
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0].set_title("gal_gadot_sandp")
ax[1].imshow(cv.cvtColor(median_blur_1, cv.COLOR_BGR2RGB))
ax[1].set_title("Median_Blur 1")
ax[2].imshow(cv.cvtColor(median_blur_2, cv.COLOR_BGR2RGB))
ax[2].set_title("Median_Blur 2")
plt.show()
```

```python
im = cv.imread("tom_dark.jpg")

# Brightness and contrast adjustments,
# im_light(i,j) = alpha * im_dark(i,j) + beta

im_light_1 = cv.convertScaleAbs(im, alpha=1, beta=30)
im_light_2 = cv.convertScaleAbs(im, alpha=1, beta=50)

#plotting the 3 images

fig, ax = plt.subplots(1,3)
fig.set_figwidth(20)
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0].set_title("tom_dark")
ax[1].imshow(cv.cvtColor(im_light_1, cv.COLOR_BGR2RGB))
ax[1].set_title("Light 1")
ax[2].imshow(cv.cvtColor(im_light_2, cv.COLOR_BGR2RGB))
ax[2].set_title("Light 2")
plt.show()
```
